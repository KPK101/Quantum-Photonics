```python
import numpy as np
import matplotlib.pyplot as plt
import qutip as qt
```

```python
a = qt.destroy(10)
x = a.dag() + a
H = a.dag()*a
times = np.linspace(0,10.0,200)
```

```python
corr1 = qt.correlation_2op_1t(H, None, times, [np.sqrt(0.5) * a], x, x)
corr2 = qt.correlation_2op_1t(H, None, times, [np.sqrt(1.0) * a], x, x)
corr3 = qt.correlation_2op_1t(H, None, times, [np.sqrt(2.0) * a], x, x)

plt.figure()
plt.plot(times, np.real(corr1), times, np.real(corr2), times, np.real(corr3))
plt.legend(['0.5','1.0','2.0'])
plt.xlabel(r'Time $t$')
plt.ylabel(r'Correlation $\left<x(t)x(0)\right>$')
plt.show()
```
