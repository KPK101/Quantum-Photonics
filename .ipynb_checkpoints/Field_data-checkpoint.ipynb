# Loading field data

```python
import numpy as np
filepath = './data/ssp_data.mat'
data = {}
f = h5py.File(filepath)
for k, v in f.items():
    data[k] = np.array(v)
```

```python
print(list(data.keys()))
```

```python
x = np.array(data['x']).reshape(-1,1)
y = np.array(data['y']).reshape(-1,1)
    
plt.plot(x)
plt.title('x')
plt.show()
plt.plot(y)
plt.title('y')
plt.show()

# x data is linearly stored; can analyse directly
# y has to be linearised; density varies in between 
```

```python
# Functions to extract data
def complex_mattopy(matdata):
    shape = np.shape(matdata)
    pydata = np.zeros(shape,dtype=complex)
    for i in range(shape[0]):
        for j in range(shape[1]):
            pydata[i][j] = matdata[i][j][0] + 1j*matdata[i][j][1]
    return pydata
def get_fields(data):
    Ex = complex_mattopy(data['Ex'])
    Ey = complex_mattopy(data['Ey'])
    return Ex,Ey
def real_(e):
    shape = np.shape(e)
    pydata = np.zeros(shape)
    for i in range(shape[0]):
        for j in range(shape[1]):
            pydata[i][j] = np.real(e[i][j])
    return pydata
    
```

```python
ex,ey = get_fields(data) # convert data to py format
ex = real_(ex) # get real field
ey = real_(ey)
```

```python
ex_corr = np.zeros((len(ex),len(ex)))
for i,e1 in enumerate(ex):
    for j,e2 in enumerate(ex):
        ex_corr[i][j] = float(np.corrcoef(e1,e2)[0][1])
        
indxs = [0,10,110]
for i,ec in enumerate(ex_corr):
    if(i>20):
        break
    plt.plot(ec)
plt.title('Correlation coefficients \n all field pairs)')
plt.show()

for i in indxs:
    plt.plot(ex_corr[i])
    plt.scatter(i,ex_corr[i][i],c='red')
plt.title('Correlation variation in sample indices')
plt.show()


```

```python
# Generating test data; Coherent sinusoid
w = 2*np.pi
t = np.linspace(0,2*np.pi/w,100)
x = np.sin(w*t)
taus = []
def shift(x,tau):
    return np.concatenate([x[tau:],x[0:tau]])

rxy = []
nprxy = []
for tau in range(len(t)):
    corrcoeff = np.corrcoef(x,shift(x,tau))
    rxy.append(corrcoeff[0][1])
    nprxy.append(np.correlate(x,shift(x,tau)))
    
plt.plot(w*t,x)
plt.plot(w*t,shift(x,len(t)//2))
plt.xlabel('wt')
plt.ylabel('x')
plt.title('x, x shifted by len(t)//2')
plt.show()

    
plt.plot(w*t,rxy)
# plt.plot(w*t,nprxy)
plt.xlabel('tau')
plt.ylabel('rxy')
plt.title('Correlation between x and its time shifted version')
plt.show()

print(min(rxy))
```

```python
def g2(ex):
    g2s = np.zeros((len(ex),len(ex)))
    def avg(e):
        return np.mean(e)
    for i in range(len(ex)):
        for j in range(len(ex)):
            g2s[i,j]=(avg(ex[i].conjugate()*ex[j].conjugate()*ex[i]*ex[j])/(avg(ex[i].conjugate()*ex[i])*avg(ex[j].conjugate()*ex[j])))
    return g2s

g2s = g2(ex)
f,axarr = plt.subplots(1,2)
f.set_figheight(10)
f.set_figwidth(10)
im1 = axarr[0].imshow(g2s,cmap='magma')
axarr[0].title.set_text(r'Variation of $g^2$ from {:.2f} to {:.2f}'.format(np.min(g2s),np.max(g2s)))
plt.colorbar(im1,ax=axarr[0],fraction=0.046, pad=0.04)
axarr[0].set_xlabel('t')
axarr[0].set_ylabel('t')

im2 = axarr[1].imshow(np.log(g2s),cmap='magma')
axarr[1].title.set_text(r'Log variation of $g^2$ from {:.2f} to {:.2f}'.format(np.min(g2s),np.max(g2s)))
plt.colorbar(im2,ax=axarr[1],fraction=0.046, pad=0.04)
axarr[1].set_xlabel('t')
axarr[1].set_ylabel('t')

plt.tight_layout()
plt.show()


g20=[]
for i,g in enumerate(g2s):
    g20.append(g[i])
    if g[i]<1:
        print('False')
plt.plot(g20)
plt.title(r'$g^2(0)$ at different instances')
plt.xlabel('t')
plt.ylabel(r'$g^2(0)$')
plt.text(5,10,f'min = {min(g20):.2f}')
plt.text(70,250,f'max = {max(g20):.2f}')
plt.show()
```

# Frequency Analysis of the electric field

```python
ex_ = ex.T
print(ex_.shape)
```

## Temporal spectrum

```python
limit = 100
for i,e in enumerate(ex_):
    if(i>limit):
        break
    fft1d = np.fft.fft(e)
    plt.plot(np.abs(fft1d)[:20]) # change the extreme to view local spectrum
plt.show()
for i,e in enumerate(ex_):
    if(i>limit):
        break
    fft1d = np.fft.fft(e)
    plt.plot(np.angle(fft1d))
plt.show()
```

## Spatial spectrum

```python
limit = 100
for i,e in enumerate(ex):
    if(i>limit):
        break
    fft1d = np.fft.fft(e)
    plt.plot(np.abs(fft1d)[:50])
plt.show()
for i,e in enumerate(ex):
    if(i>limit):
        break
    fft1d = np.fft.fft(e)
    plt.plot(np.angle(fft1d))
plt.show()
```
